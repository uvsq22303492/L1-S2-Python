Vous choisissez la taille d'image et le nombre de cercles que vous voulez. Mais le programme doit afficher une telle figure pour n'importe quelles valeurs choisies.


````python
import tkinter as tk
import time

create = True
couleurs = ["blue", "green", "black", "yellow", "magenta", "red"]  # Couleurs dans l'ordre
index_couleur = 0  # Index de la couleur actuelle

# Fonctions auxiliaires
def dessiner_cercles(taille):
    global index_couleur
    if taille > 0:
        x1 = (150 - taille) // 2
        y1 = (150 - taille) // 2
        x2 = x1 + taille
        y2 = y1 + taille
        couleur = couleurs[index_couleur]
        image.create_oval(x1, y1, x2, y2, outline=couleur, fill=couleur)
        fenetre.update()
        time.sleep(0.2)  # Pause de 0.2 seconde
        index_couleur = (index_couleur + 1) % len(couleurs)
        dessiner_cercles(taille - 10)

# Création de la fenêtre
fenetre = tk.Tk()
fenetre.title("Mon dessin")

# Création des Widgets
image = tk.Canvas(fenetre, background="white", width=150, height=150)

entree_couleur = tk.Entry(fenetre)
entree_couleur.insert(0, couleurs[index_couleur])  # Valeur par défaut
bouton_dessiner = tk.Button(fenetre, text="Dessiner Cercles", command=lambda: dessiner_cercles(150))

# Positionnement des Widgets
image.pack()
bouton_dessiner.pack()

# Lancement de la boucle
fenetre.mainloop()
````

Les questions qui suivent sont à traiter dans un fichier `couleurs.py`.

On rappelle que les couleurs sont codées par leur 3 composantes rouge, vert et bleu auxquelles on associe des valeurs comprises entre 0 et 255. La fonction suivante retourne une couleur dans le format attendu par Python à partir de ces 3 composantes.


````python
def get_color(r, g, b):
    """ Retourne une couleur à partir de ses composantes r, g, b entre 0 et 255"""
    return '#{:02x}{:02x}{:02x}'.format(r, g, b)
````

3. Copier le contenu du fichier `clic2.py` dans le fichier `clic3.py`. Modifier le fichier `clic3.py` de manière à faire les choses suivantes: 
    * afficher une ligne verticale en blanc qui sépare l'écran en 2 parties égales
    * à chaque fois que l'utilisateur a fait 2 clics, s'ils sont du même côté alors afficher une ligne bleue reliant les deux points où l'utilisateur a cliqué, sinon une ligne rouge. 

````python
import tkinter as tk

last_clicks = []

def draw_line(canvas, x1, y1, x2, y2, color):
    canvas.create_line(x1, y1, x2, y2, fill=color)

def on_click(event):
    global last_clicks
    x = event.x
    y = event.y

    last_clicks.append((x, y))

    if len(last_clicks) == 2:
        x1, y1 = last_clicks[0]
        x2, y2 = last_clicks[1]

        if x1 < 250 and x2 < 250:
            draw_line(canvas, x1, y1, x2, y2, "blue")
        elif x1 >= 250 and x2 >= 250:
            draw_line(canvas, x1, y1, x2, y2, "blue")
        else:
            draw_line(canvas, x1, y1, x2, y2, "red")

        last_clicks = []

def draw_vertical_line(canvas):
    canvas.create_line(250, 0, 250, 500, fill="white")

root = tk.Tk()
root.title("Clic")
canvas = tk.Canvas(root, width=500, height=500, bg="black")
canvas.pack()

draw_vertical_line(canvas)
canvas.bind("<Button-1>", on_click)

root.mainloop()
````

5. Copier le contenu du fichier `clic4.py` dans le fichier `clic5.py`. Modifier le fichier `clic5.py` de manière à faire les choses suivantes: 
    * pour les 8 premiers clics de l'utilisateur, on affiche un cercle rouge centré sur le point cliqué par l'utilisateur, et de rayon 50
    * au neuvième clic, tous les cercles deviennent jaunes
    * au dizième clic, tous les cercles disparaissent et on recommence au début

````python
import tkinter as tk

click_count = 0
circles = []

def draw_circle(canvas, x, y, color):
    canvas.create_oval(x - 50, y - 50, x + 50, y + 50, fill=color)

def on_click(event):
    global click_count, circles

    if click_count < 8:
        draw_circle(canvas, event.x, event.y, "red")
        circles.append("red")
    elif click_count == 8:
        for i in range(len(circles)):
            draw_circle(canvas, event.x, event.y, "yellow")
            circles[i] = "yellow"
    elif click_count == 9:
        for i in range(len(circles)):
            canvas.delete(circles[i])
        circles.clear()
        click_count = -1

    click_count += 1

root = tk.Tk()
root.title("Clic")
canvas = tk.Canvas(root, width=500, height=500, bg="white")
canvas.pack()

canvas.bind("<Button-1>", on_click)

root.mainloop()
````
